```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, a, c):
    return 1 / (1 + np.exp(-a * (x - c)))

x = np.linspace(-5, 5, 100)  # Generate a range of x values
a = 1  # Sigmoid slope parameter(shib)
c = 0  # Sigmoid center parameter(markaz)

y = sigmoid(x, a, c)  # Compute the sigmoid values

plt.plot(x, y)  # Plot the sigmoid curve
plt.xlabel('x')
plt.ylabel('Membership')
plt.title('Sigmoid Membership Function')
plt.grid(True)
plt.show()
```
